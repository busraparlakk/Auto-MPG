```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("ozcangundes/mt5-small-turkish-summarization")

model = AutoModelForSeq2SeqLM.from_pretrained("ozcangundes/mt5-small-turkish-summarization")

def generate_summary(main_news):
  source_encoding=tokenizer(
    main_news,
    max_length=784,
    padding="max_length",
    truncation=True,
    return_attention_mask=True,
    add_special_tokens=True,
    return_tensors="pt")

  generated_ids=model.generate(
      input_ids=source_encoding["input_ids"],
      attention_mask=source_encoding["attention_mask"],
      num_beams=2,
      max_length=120,
      repetition_penalty=2.5,
      length_penalty=2.0,
      early_stopping=True,
      use_cache=True
  )

  preds=[tokenizer.decode(gen_id, skip_special_tokens=True, clean_up_tokenization_spaces=True)
         for gen_id in generated_ids]

  return "".join(preds)
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

# Model ve tokenizer yükleme
model_name = "Marzu39/bert-turkish-text-classification"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

def classify_text(text):
    # Tokenizer ve model çalıştırma
    inputs = tokenizer(text, return_tensors="pt")
    outputs = model(**inputs)

    # Tahmini sınıf
    predicted_class = torch.argmax(outputs.logits, dim=1).item()
    return predicted_class

# Kullanım
text = "idare eder diyebilirim"
predicted_class = classify_text(text)
print(f"Predicted Class ID: {predicted_class}")

```

### Metin Özetleme (Text Summarization) Nedir?

Metin özetleme, uzun bir metni ana anlamını koruyarak daha kısa bir versiyona dönüştürme işlemidir. Bu, bilgiyi hızlı bir şekilde kavramak veya metin içinde önemli noktaları bulmak için çok kullanışlıdır.

### `generate_summary` Fonksiyonu Ne Yapar?

`generate_summary` fonksiyonu, Türkçe metinleri özetlemek için özel olarak eğitilmiş bir Transformer modelini (ozcangundes/mt5-small-turkish-summarization) kullanır.

İşleyişi adımları şunlardır:

1.  **Metni Token'lara Ayırma (Tokenization):** `tokenizer` (kelimeleri veya kelime parçalarını sayılara dönüştüren araç) verilen `main_news` metnini modelin anlayabileceği bir formata çevirir. Bu aşamada metin belirli bir uzunluğa (`max_length=784`) kadar kırpılır (`truncation=True`) ve sabit bir uzunluğa (`padding="max_length"`) doldurulur.

2.  **Özet Oluşturma (Generation):** `model.generate()` fonksiyonu, token'lara ayrılmış metni alarak bir özet oluşturur. Burada bazı parametreler kullanılır:
    *   `num_beams=2`: Özet oluşturulurken daha kaliteli sonuçlar elde etmek için arama stratejisini belirler.
    *   `max_length=120`: Oluşturulacak özetin maksimum uzunluğunu sınırlar.
    *   `repetition_penalty=2.5`: Modelin aynı kelimeleri veya kelime gruplarını tekrar etmesini engeller.
    *   `length_penalty=2.0`: Uzun özetlerin üretilmesini teşvik eder.
    *   `early_stopping=True`: Modelin daha kısa ve anlamlı bir özet bulduğunda durmasını sağlar.

3.  **Token'ları Metne Çevirme (Decoding):** Oluşturulan özetin sayısal temsili (ID'ler) tekrar okunabilir bir metne dönüştürülür (`tokenizer.decode`). `skip_special_tokens=True` ve `clean_up_tokenization_spaces=True` parametreleri, özetin temiz ve okunur olmasını sağlar.

Kısacası, bu fonksiyon girilen uzun bir Türkçe metni alarak, içeriğini koruyan ve daha anlaşılır olan kısa bir özetini sunar.
